### Redimensionar y pasar la imagen a escala de grises

```python
import cv2
import numpy as np
import os

path = './imagenes/'
save_path = './imagenes2/'

clasificador = cv2.CascadeClassifier("Clasificadores/haarcascade_frontalface_default.xml")
img2 = None
j = 0
for i in os.listdir(path):
    clase = i;
    if not os.path.exists(os.path.join(save_path , i)):
        os.makedirs(os.path.join(save_path , i))
    for f in os.listdir(os.path.join(path, i)):
        file_name = os.path.join(path, i,f)
        
        #Se carga la imagen en escala de grises
        img = cv2.imread(file_name,cv2.IMREAD_GRAYSCALE) 
        
        #clasificador para encontrar las posibles caras
        face = clasificador.detectMultiScale(img, scaleFactor=1.1, minNeighbors=10, minSize=(5, 5), flags=cv2.CASCADE_SCALE_IMAGE)
        if(len(face) > 0):
            for (x, y, w, h) in face: 
                img = img[y:y+h, x:x+w] 

                #Redimension de la imagen a 200x200
                img2 = cv2.resize(img, (290, 200)) 
                file_save = os.path.join(save_path, clase,'imagen')
                cv2.imwrite("{}{}.png".format(file_save,j), img2)
                print('Gardada {} => {}{}.png'.format(file_name,file_save,j))
                j += 1
```
